# Kaggle Ames House Prices

This project aims to create some regression models to predict the sale price of each home in Ames, Lowa. 
The dataset was collected from Kaggle competition website (https://www.kaggle.com/c/house-prices-advanced-regression-techniques), which includes 1460 observations and 81 variables that cover all aspects of residential homes. The complete code can be found [here](https://github.com/Erica-Tan/House-Prices).

we are interested to know:
•	Which neighborhood is most popular to live in? 
•	Which neighborhood in Ames has the most expensive home of all?
•	Which residential home has the best value for money? 
•	What are the contributing factors to soaring housing price?

The output of the solution will give useful insights to any residential selling business in terms of market trend, price planning, and customized marketing and selling. 

Before we get started, let’s first import all of the functions and classes we intend to use.

```python
# Load data
df_data = pd.read_csv("data/train.csv")
```

## 1. Missing data

This dataset is filled with many missing values, therefore, all NA’s should be filled in with appropriate values before building any predictive model. For each column, the missing values were replaced by using features that have a strong correlation, which helps to determine what values to fill in.

```python
missing  = df_data.isnull().sum()
missing = missing[missing > 0]
total  = missing.sort_values(ascending=False)
percent = (missing/len(df_data)).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=["Total", "Percent"])

print(missing_data)

#plot missing data
total.plot.bar()
plt.show()
```

Let's define a helper function to fill with missing data first.

```python
def fill_missing_data(frame, column, value = 'None', re_column='None'):
    filtered = frame[column].isnull()
    if re_column == 'None':
        frame.loc[filtered, column] = value
    else:
        frame.loc[filtered, column] = frame.loc[filtered, re_column]
```

### PoolQC
PoolQC can be filled with ‘None’ since each has a corresponding PoolArea of 0

```python
filtered = (df_data["PoolQC"].isnull())
df_data.loc[filtered,["PoolQC","PoolArea"]]
```

```python
fill_missing_data(df_data, "PoolQC", "None")
```

### GarageType / GarageYrBlt / GarageFinish / GarageCars / GarageArea / GarageQual / GarageCond

For the garage attributes, we can check to see that if the missing recordes have 0 GarageArea and 0 GarageCars. If they do, we can fill in them with 'None' or '0' since having 0 area and 0 cars in their garage will imply that they do not have any at all.

```python
garage_cols=["GarageType","GarageYrBlt","GarageFinish","GarageCars","GarageArea","GarageQual","GarageCond"]

filtered = df_data[garage_cols].isnull().apply(lambda x: any(x), axis=1)

df_data.loc[filtered, garage_cols]
```

```python
for column in garage_cols:
    if df_data.dtypes[column] == "object":
        fill_missing_data(df_data, column, "None")
    else:
        fill_missing_data(df_data, column, 0)
```

### Electrical

With only 1 missing value for Electrical. we can fill in it with the most frequent value 'SBrkr'.

```python
sns.countplot(df_data["Electrical"])
plt.show()
```

```python
# most frequent value 
fill_missing_data(df_data, "Electrical", "SBrkr")
```

### BsmtQual / BsmtCond / BsmtExposure / BsmtFinType1 / BsmtFinType2

The table below shows that there are 11 basement features and each with at least 1 missing value. 

```python
basement_cols=["BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","BsmtFinSF1","BsmtFinSF2"]

filtered = df_data[basement_cols].isnull().apply(lambda x: any(x), axis=1)

df_data.loc[filtered, basement_cols]
```

For rows 948, the only missing value is from BsmtExposure attribute, we can fill in 'No' as that is the most frequent value and this house most likely don’t have any exposure for its basement.

```python
df_data.loc[948, "BsmtExposure"] = "No"
```

For Rows 332, the only missing value is from BsmtFinType2 attribute and we should fill the missing value based on BsmtFinSF2.

```python
grouped = df_data.groupby("BsmtFinType2")
grouped = grouped["BsmtFinSF2"].agg(np.mean)
print(grouped)

df_data.loc[332, "BsmtFinType2"] = "LwQ"
```

The rest of the basement columns corresponding to area will be filled with '0' since they likely don’t have a basement and the categoric missing values will be filled with 'None'.

```python
for column in basement_cols:
    if df_data.dtypes[column] == "object":
        fill_missing_data(df_data, column, "None")
    else:
        fill_missing_data(df_data, column, 0)
```

### MasVnrType / MasVnrArea

We can see if both missing values come from the same houses. For houses with NA’s on both columns, we can fill '0' for the area and 'None' for the type since they likely do not have a masonry veneer.

```python
filtered = (df_data["MasVnrType"].isnull()) | (df_data["MasVnrArea"].isnull())
df_data.loc[filtered,["MasVnrType","MasVnrArea"]]
```

```python
fill_missing_data(df_data, "MasVnrType", "None")
fill_missing_data(df_data, "MasVnrArea", 0)
```

### LotFrontage

We can group each neighborhood and take the median of each LotFrontage and fill the missing values of each LotFrontage based on what neighborhood the house comes from.

```python
grouped = df_data.groupby("Neighborhood")
grouped = grouped["LotFrontage"].agg(np.mean)
print(grouped)
```

```python
filtered = df_data["LotFrontage"].isnull()
df_data.loc[filtered,"LotFrontage"] = df_data.loc[filtered,"Neighborhood"].map(lambda neighbor : grouped[neighbor])
```

### Fence / MiscFeature
The missing values for Fence and MiscFeature can be repalced with ‘None’ as they probably don’t have this feature with their property.

```python
fill_missing_data(df_data, "Fence", "None")
fill_missing_data(df_data, "MiscFeature", "None")
```

### FireplaceQu

Let's check to see if any of the missing values for FireplaceQu come from houses that recorded having at least 1 fireplace

```python
filtered = (df_data["Fireplaces"] > 0) & (df_data["FireplaceQu"].isnull())
len(df_data.loc[filtered,["Fireplaces","FireplaceQu"]])
```

All the houses that have missing values did not recorded having any fireplaces. We can simply replace the NA’s with ‘None’ since these houses don’t have any fireplaces at all.

```python
fill_missing_data(df_data, "FireplaceQu", "None")
```

### Alley
We can fill ‘NA’ for any of the houses with None’s as these houses may not have any type of alley access.

```python
fill_missing_data(df_data, "Alley", "None")
```

## 2. Feature construction

```python
#split data into two train and test
train, test = train_test_split(df_data, test_size = 0.2, random_state=42)
```

Some numerical features are actually really categories

```python
df_data = df_data.replace({"MSSubClass" : {20 : "SC20", 30 : "SC30", 40 : "SC40", 45 : "SC45", 
                                       50 : "SC50", 60 : "SC60", 70 : "SC70", 75 : "SC75", 
                                       80 : "SC80", 85 : "SC85", 90 : "SC90", 120 : "SC120", 
                                       150 : "SC150", 160 : "SC160", 180 : "SC180", 190 : "SC190"},
                           "OverallQual" : {10 : "Ve", 9 : "Ex", 8 : "Vg" , 7 : "Gd", 6 : "AA",
                                        5 : "TA", 4 : "BA" , 3 : "Fa", 2 : "Po", 1 : "Vp"},
                           "OverallCond" : {10 : "Ve", 9 : "Ex", 8 : "Vg" , 7 : "Gd", 6 : "AA",
                                        5 : "TA", 4 : "BA" , 3 : "Fa", 2 : "Po", 1 : "Vp"}
                           })
```

### Year / Month

```python
# records that a house was remodelled if the year it was built is different than the remodel year.
df_data["Remodeled"] = (df_data["YearRemodAdd"] != df_data["YearBuilt"]) * 1

# if the houses have been recently remodelled 
df_data["RecentRemodel"] = (df_data["YearRemodAdd"] == df_data["YrSold"]) * 1
    
# Was this house sold in the year it was built?
df_data["NewHouse"] = (df_data["YearBuilt"] == df_data["YrSold"]) * 1

# how old a house is
df_data["HouseAge"] = date.today().year - df_data["YearBuilt"]

# how long ago the house was sold
df_data["TimeSinceSold"] = date.today().year - df_data["YrSold"]

# how many years since the house was remodelled and sold 
df_data["YearSinceRemodel"] = df_data["YrSold"] - df_data["YearRemodAdd"]
```

The following graph shows that the largest proportion of houses sold is during the summer months: May, June, July. Let’s add a column that seperates the the summer houses from the rest.

```python
sns.countplot(train["MoSold"])
plt.show()

df_data["HighSeason"] = (df_data["MoSold"].isin([5,6,7])) * 1
```

### Area features

```python
area_columns = ["LotFrontage", "LotArea", "MasVnrArea", "BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF",
             "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "GrLivArea", "GarageArea", "WoodDeckSF", 
             "OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch", "LowQualFinSF", "PoolArea"]

df_data["TotalArea"] = df_data[area_columns].sum(axis =1)


df_data["HasLowQualArea"] = (df_data["LowQualFinSF"] > 0) * 1
df_data["HasBasement"] = (df_data["TotalBsmtSF"] > 0) * 1
df_data["HasGarage"] = (df_data["GarageArea"] > 0) * 1
df_data["Has2ndFloor"] = (df_data["2ndFlrSF"] > 0) * 1
df_data["HasMasVnr"] = (df_data["MasVnrArea"] > 0) * 1
df_data["HasWoodDeck"] = (df_data["WoodDeckSF"] > 0) * 1
df_data["HasOpenPorch"] = (df_data["OpenPorchSF"] > 0) * 1
df_data["HasScreenPorch"] = (df_data["ScreenPorch"] > 0) * 1
df_data["HasEnclosedPorch"] = (df_data["EnclosedPorch"] > 0) * 1
df_data["Has3SsnPorch"] = (df_data["3SsnPorch"] > 0) * 1
df_data["HasPool"] = (df_data["PoolArea"] > 0) * 1
```

### Neighborhood

```python
grouped = grouped = train[["Neighborhood", 'SalePrice']].groupby("Neighborhood").mean()
grouped = grouped.sort_values(by='SalePrice', ascending=0)
grouped = grouped.reset_index(level=0)

sns.barplot(grouped["Neighborhood"], grouped["SalePrice"])
plt.xticks(rotation=30)
plt.show()

#one-hot encode the more expensive neighborhoods 
nbrh_rich = ["NridgeHt", "NoRidge", "StoneBr", "Veenker", "Timber", "Somerst"]
df_data["NbrhRich"] = df_data["Neighborhood"].apply(lambda x: 1 if x in nbrh_rich else 0)
```

### SalePrice

The distribution of SalePrice is skewed. So it will be transformed later

```python
#descriptive statistics summary
print (train["SalePrice"].describe())

#histogram
# 1. Deviate from the normal distribution.
# 2. Have appreciable positive skewness.
# 3. Show peakedness.
sns.distplot(train["SalePrice"], fit=stats.norm)
plt.show()

#normal probability plot
res = stats.probplot(train["SalePrice"], plot=plt)
plt.show()

#skewness and kurtosis
print("Skewness: %f" % train["SalePrice"].skew())
print("Kurtosis: %f" % train["SalePrice"].kurt())
```

### Remove columns

Utilities only has 1 value for 'NoSeWa' and the rest are 'AllPub'. We can drop this feature from our dataset as the house with 'NoSeWa' is from our training set and will not have any effect on predictive modelling.

```python
sns.countplot(df_data["Utilities"])
plt.show()
```

For date attribures, since we have constructed some features based on these features, we removed the original attributes. Garage area is strongly correlated with number of cars, so we can remove 'GarageCars'. "1stFlrSF" amd "2ndFlrSF attributes can also be removed as 1stFlrSF + 2nFlrSF + LowQualFinSF = GrLivArea. 

```python
del_columns = ["Utilities", "GarageCars", "1stFlrSF", "2ndFlrSF", "YearBuilt",
               "YearRemodAdd", "YrSold", "MoSold", "GarageYrBlt", "Id"]

df_data = df_data.drop(del_columns, axis=1)
```

## 3. Dummy encoding

Before performing dummy, numeric columns are recorded for later use.

```python
num_columns = df_data.select_dtypes(include = ["float64", "int64"]).columns.values.tolist()
num_columns.remove("SalePrice")

# one hot encoding for categorical data
df_data = pd.get_dummies(df_data)
```

## 4. Log transform

Transforming the skewed numeric features by taking log(feature + 1) will make the features more normal. 

Some independent variables look like good candidates for log transformation, such asTotalBsmtSF, KitchenAbvGr, LotFrontage, LotArea and others. While ganining on regression transformation will smooth out some irregularities which could be important, such as 2ndFlrSF. These irregularities are good candidates for feature construction.

```python
# log the attributes that follow the normal distribute
log_columns = ["LotFrontage", "LotArea","BsmtUnfSF",
               "TotalBsmtSF", "GrLivArea", "TotalArea", "SalePrice"]

# Transform the skewed numeric features by taking log(feature + 1).
# This will make the features more normal.
skewed = df_data[log_columns].apply(lambda x: x.dropna().astype(float).skew())
skewed = skewed[(skewed > 0.8) | (skewed < -0.8)]
skewed = skewed.index.tolist()

pd.options.mode.chained_assignment = None 
df_data.loc[:,skewed] = np.log1p(df_data[skewed].values)
```

## 4. Normalization

An important concern with this dataset is that the input attributes all vary in their scales because they measure different quantities. It is almost always good practice to normalized the data before buliding model.

Normalization cannot be done before the splitting as we don't want to fit the MinMaxScaler on some observations that will later be used in the test set.

```python
# split the dataset into dataset 80% and test 20%
train, test = train_test_split(df_data, test_size = 0.2, random_state=42)

min_max = MinMaxScaler()
train.loc[:,num_columns] = min_max.fit_transform(train[num_columns])
test.loc[:,num_columns] = min_max.fit_transform(test[num_columns])
```

## 5. Build model - XGBoost 

```python
df_data = train

# split train dataset into train and validation
train, validation = train_test_split(train, test_size=0.2, random_state=42)

# split into input (X) and output (Y) variables
y_train = train['SalePrice']
x_train = train.drop('SalePrice', axis=1)
y_validation = validation['SalePrice']
x_validation = validation.drop('SalePrice', axis=1)
Y = df_data['SalePrice']
X = df_data.drop('SalePrice', axis=1)
```

### Baseline model

Baseline model provides a point of comparison for the model that will be evaluated later.  We can also specify an evaluation metric and set the early_stopping_rounds parameter to stop training once no further improvements have been made to the model.

```python
# Baseline model
model = xgb.XGBRegressor(
     learning_rate =0.1,
     n_estimators=1000,
     max_depth=5,
     min_child_weight=2,
     gamma=0,
     subsample=0.8,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     random_state=42)

early_stopping_rounds = 10

#Fit the algorithm on the data
eval_set = [(x_validation, y_validation)]
model.fit(x_train, y_train, early_stopping_rounds=early_stopping_rounds, eval_metric='rmse', eval_set=eval_set, verbose=False)

# make predictions for train data
y_train_pred = model.predict(x_train)
y_train_pred = [value for value in y_train_pred]
# evaluate predictions
y_train_true = np.expm1(y_train.values)
y_train_pred = np.expm1(y_train_pred)
print("train RMSE: %f" % (np.sqrt(mean_squared_error(y_train_true, y_train_pred))))

# make predictions for validation data
y_validation_pred = model.predict(x_validation)
y_validation_pred = [value for value in y_validation_pred]
# evaluate predictions
y_validation_true = np.expm1(y_validation.values)
y_validation_pred = np.expm1(y_validation_pred)
print("validation RMSE: %f" % (np.sqrt(mean_squared_error(y_validation_true, y_validation_pred))))
```

### Feature importance

Feature importance provides a score that indicates how useful or valuable each feature was in the construction of the boosted decision trees within the model. The more an attribute is used to make key decisions with decision trees, the higher its relative importance.

```python
#Plot important
xgb.plot_importance(model, max_num_features = 20)
plt.show()
```

### Parameter tuning
Step 1: Tune learning rate and number of estimators

```python
#Tune learning rate and number of estimators
model = xgb.XGBRegressor(
     learning_rate =0.1,
     n_estimators=1000,
     max_depth=5,
     min_child_weight=2,
     gamma=0,
     subsample=0.8,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     random_state=42)

# define the grid search parameters
# best parameters {'n_estimators': 250 / 150, 'learning_rate': 0.1}
learning_rate = [0.01, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4]
n_estimators = range(50, 400, 50)

param_test = {
 'learning_rate': learning_rate,
  'n_estimators': n_estimators
}

gsearch = GridSearchCV(estimator=model, scoring="neg_mean_squared_error", n_jobs=-1,
                        param_grid=param_test, verbose=0)
gsearch.fit(x_train, y_train)

# summarize results
print("Best: %f using %s" % (gsearch.best_score_, gsearch.best_params_))
```

Step 3: Tune max_depth and min_child_weight

```python
# Tune max_depth and min_child_weight
model = xgb.XGBRegressor(
     learning_rate=0.1,
     n_estimators=150,
     max_depth=5,
     min_child_weight=2,
     gamma=0,
     subsample=0.8,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     random_state=42)

# define the grid search parameters
# best parameters {'min_child_weight': 5, 'max_depth': 5}
param_test = {
    'max_depth':range(3,10,2),
    'min_child_weight':range(1,6,2)
}

gsearch = GridSearchCV(estimator=model, scoring="neg_mean_squared_error", n_jobs=-1,
                        param_grid=param_test, verbose=0)
gsearch.fit(x_train, y_train)

# summarize results
print("Best: %f using %s" % (gsearch.best_score_, gsearch.best_params_))
```

Step 3: Tune gamma

```python
# Tune gamma
model = xgb.XGBRegressor(
     learning_rate=0.1,
     n_estimators=150,
     max_depth=5,
     min_child_weight=5,
     gamma=0,
     subsample=0.8,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     random_state=42)

# define the grid search parameters
# best parameters {'gamma': 0.0}
param_test = {
 'gamma':[i/10.0 for i in range(0,5)]
}

gsearch = GridSearchCV(estimator=model, scoring="neg_mean_squared_error", n_jobs=-1,
                        param_grid=param_test, verbose=0)
gsearch.fit(x_train, y_train)

# summarize results
print("Best: %f using %s" % (gsearch.best_score_, gsearch.best_params_))
```

Step 4: Tune subsample and colsample_bytree

```python
# Tune subsample and colsample_bytree
model = xgb.XGBRegressor(
     learning_rate=0.1,
     n_estimators=150,
     max_depth=5,
     min_child_weight=5,
     gamma=0,
     subsample=0.8,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     random_state=42)

# define the grid search parameters
# best parameters {'colsample_bytree': 0.6, 'subsample': 0.8}
subsample = [i/100.0 for i in range(10,110,10)]
colsample_bytree = [i/100.0 for i in range(10,110,10)]

param_test = {
 'subsample': subsample,
 'colsample_bytree':colsample_bytree
}

gsearch = GridSearchCV(estimator=model, scoring="neg_mean_squared_error", n_jobs=-1,
                        param_grid=param_test, verbose=0)
gsearch.fit(x_train, y_train)

# summarize results
print("Best: %f using %s" % (gsearch.best_score_, gsearch.best_params_))
```

Step 5: Tuning Regularization Parameters

```python
# Tuning Regularization Parameters
model = xgb.XGBRegressor(
     learning_rate=0.1,
     n_estimators=150,
     max_depth=5,
     min_child_weight=5,
     gamma=0,
     subsample=0.6,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     random_state=42)

# define the grid search parameters
# best parameters {'reg_alpha': 0.005}
param_test = {
  'reg_alpha':[0, 0.001, 0.005, 0.01, 0.05]
}

gsearch = GridSearchCV(estimator=model, scoring="neg_mean_squared_error", n_jobs=-1,
                        param_grid=param_test, verbose=0)
gsearch.fit(x_train,y_train)

# summarize results
print("Best: %f using %s" % (gsearch.best_score_, gsearch.best_params_))
```

### Bias-Variance trade-off visulization

We can retrieve the performance of the model on the evaluation dataset and plot it to get insight into how learning unfolded while training.

```python
model = xgb.XGBRegressor(
     learning_rate=0.1,
     n_estimators=150,
     max_depth=5,
     min_child_weight=5,
     gamma=0,
     subsample=0.6,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     reg_alpha=0.005,
     random_state=42)

early_stopping_rounds = 10

#Fit the algorithm on the data
eval_set = [(x_train, y_train),(x_validation, y_validation)]
model.fit(x_train, y_train, early_stopping_rounds=early_stopping_rounds, eval_metric='rmse', eval_set=eval_set, verbose=False)


# make predictions for test data
y_train_pred = model.predict(x_train)
y_test_pred = [value for value in y_train_pred]
# evaluate predictions
y_train_real = np.expm1(y_train.values)
y_train_pred = np.expm1(y_train_pred)
print("train RMSE: %f" % (np.sqrt(mean_squared_error(y_train_real, y_train_pred))))

# make predictions for test data
y_validation_pred = model.predict(x_validation)
y_validation_pred = [value for value in y_validation_pred]
# evaluate predictions
y_validation_real = np.expm1(y_validation.values)
y_validation_pred = np.expm1(y_validation_pred)
print("validation RMSE: %f" % (np.sqrt(mean_squared_error(y_validation_real, y_validation_pred))))


results = model.evals_result()

# retrieve performance metrics
epochs = len(results['validation_0']['rmse'])
x_axis = range(0, epochs)

# plot curve
fig, ax = plt.subplots()
ax.plot(x_axis, results['validation_0']['rmse'], label='Train')
ax.plot(x_axis, results['validation_1']['rmse'], label='Validation')
ax.legend()
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.title('Validation Curve for XGBoost')
plt.show()
```

### Final model

Finalizing the model involves fitting the final model on on the entire train dataset and saving the model for later use.

```python
model = xgb.XGBRegressor(
     learning_rate=0.1,
     n_estimators=150,
     max_depth=5,
     min_child_weight=5,
     gamma=0,
     subsample=0.6,
     colsample_bytree=0.8,
     n_jobs=2,
     scale_pos_weight=1,
     reg_alpha=0.005,
     random_state=42)

# Save final model (combine train and validation)
model.fit(X, Y, eval_metric='rmse')

# save model to file
joblib.dump(model, "output/xgboost.dat")
```

### Model validation

```python
# load model from file
loaded_model = joblib.load("output/xgboost.dat")

# load test data
test = pd.read_csv("processed/test+dummy+log+minmax.csv")

y_test = test['SalePrice']
x_test = test.drop('SalePrice', axis=1)

# make predictions for test data
y_test_pred = loaded_model.predict(x_test)
y_test_pred = [value for value in y_test_pred]
# evaluate predictions
y_test_real = np.expm1(y_test.values)
y_test_pred = np.expm1(y_test_pred)
print("test RMSE: %f" % (np.sqrt(mean_squared_error(y_test_real, y_test_pred))))
```
